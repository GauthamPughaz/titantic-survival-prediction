```
def scale_features(df, features):
  n = df.shape[0]
  for feature in features:
    df[feature] -= df[feature].mean()
    std = (df[feature]**2).sum() / n
    df[feature] /= std
    
  return df
```

```
def clean_data(df):
  print("Cleaning data - STARTED")
  
  print("Dropping non-productive columns")
  df = df.drop(["PassengerId", "Name", "Ticket"], axis=1)
  
  print("Handling NaN columns")
  df["Age"] = df["Age"].fillna(df["Age"].mean())
  df["Fare"] = df["Fare"].fillna(df["Fare"].mean())
  df["Cabin"] = df["Cabin"].fillna("NA")
  df["Embarked"] = df["Embarked"].fillna("NA")
  
  print("Scaling features with respect to their mean and standard deviation")
  df = scale_features(df, ["Age"])
  
  print("Scaling down Fare, SibSp and Parch")
  df["Fare"] /= 1000
  df["SibSp"] /= 10
  df["Parch"] /= 10
  
  print("Handling categorical columns")
  oh_encoder = ce.OneHotEncoder(cols=["Sex", "Pclass", "Embarked"])
  df = oh_encoder.fit_transform(df)
  bin_encoder = ce.BinaryEncoder(cols=["Cabin"])
  df = bin_encoder.fit_transform(df)
  df = df.drop("Cabin_0", axis=1)
  
  print("Cleaning data - COMPLETED")
  return df
```

```
df = df_train.append(df_test, ignore_index=True, sort=False)
df_cleaned = clean_data(df)
```
